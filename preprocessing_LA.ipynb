## Preprocessing

```python
subsector_mapping = {
    'Agriculture':                          'agrc',
    'Agriculture|Soil Management':          'soil',
    'Carbon Capture and Sequestration':     'ccsq',
    'Energy Technology':                    'entc',
    'Energy Technology|Fugitive Emissions': 'fgtv',
    'Energy Technology|Industrial Energy':  'inen',
    'Forest':                               'frst',
    'Fugitive Emissions':                   'fgtv',
    'IPPU':                                 'ippu',
    'Land Use':                             'lndu',
    'Livestock':                            'lvst',
    'Livestock Manure Management':         'lsmm',
    'Soil Management':                      'soil',
    'Solid Waste':                          'waso',
    'Stationary Combustion and Other Energy':'scoe',
    'Transportation':                       'trns',
    'Wastewater Treatment':                 'trww',
}

subsector_mapping_df = ppu.subsector_dict_to_df(subsector_mapping)
subsector_mapping_df
```

```python
louisiana_inv_df_mapped = pd.merge(louisiana_inv_df, subsector_mapping_df, left_on="subsector", right_on="la_subsector", how="left")
# Check for missing values in code column
missing_codes = louisiana_inv_df_mapped[louisiana_inv_df_mapped["ssp_subsector"].isna()]
print(f"Missing codes in the mapping: {missing_codes.shape[0]} rows")
louisiana_inv_df_mapped.head()
```

```python
fields_to_remove = [
    "variable_field",
    "variables",
    "accounted",
    "sit_units",
    "sit_gas_measure",
    "sit_sheet",
    "sit_sections",
    "sit_subsections",
    "sit_categories",
    "sit_rows",
    "notes"
]
```

```python
# Drop subsector, la_subsector columns and rename ssp_subsector to subsector and move it to the front
louisiana_inv_df_mapped = louisiana_inv_df_mapped.drop(columns=["subsector", "la_subsector"])
louisiana_inv_df_mapped = louisiana_inv_df_mapped.rename(columns={"ssp_subsector": "subsector"})
subsector_col = louisiana_inv_df_mapped.pop("subsector")
louisiana_inv_df_mapped.insert(0, "subsector", subsector_col)
louisiana_inv_df_mapped = louisiana_inv_df_mapped.drop(columns=fields_to_remove)
louisiana_inv_df_mapped.head()
```

```python
# Group by subsector and gas, summing the values
louisiana_inv_df_groupped = louisiana_inv_df_mapped.groupby(["subsector", "gas"], as_index=False).sum()
louisiana_inv_df_groupped.head()
```

```python
louisiana_inv_df_groupped[louisiana_inv_df_groupped["subsector"] == "ippu"]
```

```python
ssp_fc_gases = ['c2f6', 'c2h3f3', 'c2hf5', 'c3f8', 'c3h2f6',
       'c3h3f5', 'c3hf7', 'c4f10', 'c4f6', 'c4h5f5', 'c5f12', 'c5f8',
       'c5h2f10', 'c6f14', 'cc4f8', 'cf4', 'ch2f2', 'ch2fchf2',
       'ch3cci2f', 'ch3ccif2', 'ch3chf2', 'ch3f', 'chf2cf3', 'chf2chf2',
       'chf3', 'hfcs', 'other_fcs', 'pfcs', 'sf6']

print("Amount of FC gases: ", len(ssp_fc_gases))

louisiana_inv_df_expanded = ppu.distribute_fc_gases(
    louisiana_inv_df_groupped,
    ssp_fc_gases
)
```

```python
# Filter the LA inventory to desired year
louisiana_inv_df_final = louisiana_inv_df_expanded[["subsector", "gas", 2021]].copy()
louisiana_inv_df_final = louisiana_inv_df_final.rename(columns={2021: "LA", "gas": "Gas", "subsector": "Subsector"})
louisiana_inv_df_final.head()
```

```python
# Merge into target format
updated_emissions_inventory_df = pd.merge(
    target_format_df,
    louisiana_inv_df_final,
    on=["Subsector", "Gas"],
    how="left"
)
```
